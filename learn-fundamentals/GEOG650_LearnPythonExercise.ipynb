### **Loops**

I learned about two types of loops available in the Python programming language: `for` and `while`.

The first was the `for` loop, which iterates over a given sequence. This loop can be used with an iterable object or function such as the list object or `range()` function. For example, consider the following code sample.

```python
# Prints out the numbers 2,3,5,7 given in a list
print("Here are a few prime numbers:")
primes = [2, 3, 5, 7]
for prime in primes: # for each number in a given list of prime numbers
    print(prime, end=' ') # print the primes on the same line

# Prints out the numbers 0,1,2,3,4
print("\nHere are the numbers up to five:")
for x in range(5): # for each number up to five
    print(x, end=' ') # print the numbers on the same line
```

The second was the `while` loop, which will repeat until a specified Boolean condition is met. For example:

```python
# Prints out the numbers 0,1,2,3,4
count = 0
while count < 5: # repeat the loop while the value of count is still less than 5
    print(count, end=' ')
    count += 1  # could also be written as count = count + 1 to increase by 1 each loop
```

### **Dictionaries**

I learned that the dictionary is a fundamental data structure in Python because it provides a quick way of retrieving a *value* associated with a *key*. Dictionaries are defined using curly braces in the general form of `{key : value}` where the key and value are linked together in such a way that the key can be used to access the value. Because the key and value are associated with one another, dictionaries are unordered collections by design. For this reason, tasks such as membership checks that check for the existence of a key in the dictionary are quite fast.

This tutorial used the example of a phonebook, which is a database of phone numbers associated with individuals. A Python dictionary can be used to build this database as follows:

```python
phonebook = {} # the "Pythonic" way of defining an empty dictionary

# assign key and value pairs using the general form of dictionary[key] = value
phonebook["John"] = 938477566
phonebook["Jack"] = 938377264
phonebook["Jill"] = 947662781

# print the contents of the "phonebook" dictionary
print(phonebook)
```

Additionally, the database could be built using a dictionary literal.

```python
# create the same phonebook using a dictionary literal
phonebook = {
    "John" : 938477566,
    "Jack" : 938377264,
    "Jill" : 947662781
}
print(phonebook)
```

One important functionality of dictionaries is that the data structure is iterable, although it is unordered. For example, the following code block demonstrates how to iterate over the "phonebook" dictionary that was previously defined in the above code.

```python
# for each name and number pair in the items of the phonebook dictionary
for name, number in phonebook.items():
    print("Phone number of %s is %d" % (name, number)) # print this formatted string
```

Perhaps one of the items in the phonebook is outdated and simply needs to be updated or removed. For this case, Python dictionaries have the ability to be changed. For example, a name can be removed from the previously defined dictionary using the `del` keyword as follows:

```python
del phonebook["John"] # delete an object from the phonebook
print(phonebook)
```

Finally, the exercise in this tutorial puts all these concepts into action by completing the following tasks.

```python
# Tutorial 11: Dictionaries

# create a dictionary using curly braces and assign it to the "phonebook" variable
# note that the keys are names and the values are phone numbers {key : value}
phonebook = {
    "John" : 938477566,
    "Jack" : 938377264,
    "Jill" : 947662781
}

# add the name "Jake" to the "phonebook" dictionary with his phone number as the value
phonebook["Jake"] = 938273443 # general form of dictionary[key] = value

# remove the name "Jill" from the "phonebook" dictionary
del phonebook["Jill"] # use the del keyword to delete an object, in this case "Jill"

# testing code to check if Jake is now in the phonebook
if "Jake" in phonebook:
    print("Jake is listed in the phonebook.") # execute if Jake is in the phonebook

# testing code to confirm that Jill is no longer in the phonebook
if "Jill" not in phonebook:
    print("Jill is not listed in the phonebook.") # execute if True
```

```python
# create a dictionary with single keys matched to multiple values
dict = {"country": ["Brazil", "Russia", "India", "China", "South Africa"],
       "capital": ["Brasilia", "Moscow", "New Dehli", "Beijing", "Pretoria"],
       "area": [8.516, 17.10, 3.286, 9.597, 1.221],
       "population": [200.4, 143.5, 1252, 1357, 52.98] }

import pandas as pd # import the Pandas module under the alias of "pd"
brics = pd.DataFrame(dict) # create a DataFrame structure from the dictionary
print(brics) # print the DataFrame
```

The default index values are numeric; however, the index values can be customized. The index values of the previously used code sample are changed below to represent the two-letter code of each country.

```python
# Set the index for the DataFrame "brics"
brics.index = ["BR", "RU", "IN", "CH", "SA"] # instead of numeric values, use the two-letter country code

# Print out the DataFrame with its new index values
print(brics)
```

Pandas provides additional functionality to index DataFrames with `iloc` and `loc` for data selection. The first operation, `iloc`, allows you to access data using an integer-based index. For example:

```python
# India is located at index position 2
# and its data are selected using "iloc" with an integer based index

print(brics.iloc[2]) # print out data, the keys and values, for India
```

The second operation, `loc`, allows you to access data using a label-based index. For example:

```python
# data for Russia and China can be accessed using "loc" with a label based index
# Pandas completes the data selection

print(brics.loc[['RU', 'CH']]) # print out data for Russia and China
```
